```python
import numpy as np

df["subtitle"] = (
    np.where(df["subtitle"].isna(), df["title"],
             df[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
df.dropna(subset=['description'],
          inplace=True
         )

df.reset_index()
```
